## Overview of the Sales Transactions Dataset

### Dataset Description
Dataset Utilized: https://archive.ics.uci.edu/dataset/292/wholesale+customers

This dataset comprises weekly sales transaction data for various products over a year. Each row represents a unique product, and each column contains sales figures for one of the 52 weeks, labeled from `W0` to `W51`. The dataset also includes normalized sales data for each week, providing a standardized measure of sales volumes relative to each product's maximum sales.

### Columns
- **Product_Code**: A unique identifier for each product.
- **W0 to W51**: Sales figures for each product from week 0 to week 51.
- **Normalized 0 to Normalized 51**: Normalized sales data from week 0 to week 51, scaled between 0 and 1 based on the maximum sales of each product.

### Purpose
The dataset is used to analyze sales patterns, forecast demand, and optimize inventory management. It serves as a fundamental resource for understanding product performance across different periods and planning marketing strategies accordingly.

## 1. Correlation Matrix Overview

The correlation matrix provided below visualizes the relationships between weekly sales across different weeks. Each cell in the matrix represents the correlation coefficient between the sales of two different weeks:

- A **positive value** (closer to +1.0) indicates a strong positive relationship, meaning as sales in one week increase, sales in the corresponding week tend to increase as well.

The color scale enhances the visualization:

This matrix is a useful tool for identifying patterns in sales over time, which can help in forecasting future trends and understanding the behavior of sales across different periods.

## 2. Distribution Analysis of Weekly Sales Data

### Purpose
This section analyzes the distribution of sales data for each week to understand how sales figures are spread over time and to identify which probability distribution fits each week's sales data best.

### Method
- **Histograms** are used to visually inspect the distribution of sales for each week. The shape of the histogram helps in identifying the type of distribution (e.g., normal, skewed, bimodal).
- **Kernel Density Estimation (KDE)** overlays a smoothed curve on the histograms to better visualize the underlying distribution.
- **Normality Test**: The D'Agostino and Pearson's test, which computes skewness and kurtosis, is applied to test if the sales data for each week follow a normal distribution.

### Interpretation
- Each plot represents a week's sales distribution, with the x-axis denoting the number of sales and the y-axis showing the frequency of those sales figures.
- The results from the normality tests provide a statistic and a p-value:
  - **Statistic**: A higher value suggests a greater deviation from normal distribution.
  - **p-value**: A value below 0.05 typically indicates that the distribution of the data is not normal.

### Conclusion
Understanding the distribution of weekly sales helps in selecting appropriate statistical models for forecasting and analysis. If the sales data for a week do not follow a normal distribution, alternative statistical techniques or transformations might be necessary to model such data accurately.


```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import normaltest
import pandas as pd

# Assuming df is your DataFrame and has been previously loaded with weekly sales data
weekly_sales_data = df.loc[:, 'W0':'W51']

def plot_distribution_and_test_normality(data, columns):
    # Adjust the layout based on the number of columns
    num_columns = len(columns)
    num_rows = (num_columns + 9) // 10  # Ensure all plots fit, assuming 10 plots per row for simplicity
    fig, axes = plt.subplots(nrows=num_rows, ncols=10, figsize=(20, 3 * num_rows))
    axes = axes.ravel()  # Flatten the axes array for easy iteration
    results = {}

    for i, col in enumerate(columns):
        # Plot histogram with kernel density estimate
        sns.histplot(data[col], bins=15, kde=True, ax=axes[i])
        axes[i].set_title(col)

        # Normality test using D'Agostino and Pearson's test
        stat, p_value = normaltest(data[col])
        results[col] = {'Statistic': stat, 'p-value': p_value}

        # Formatting
        axes[i].tick_params(axis='x', labelsize=8)
        axes[i].tick_params(axis='y', labelsize=8)
    
    for i in range(i + 1, len(axes)):  # Hide unused subplots
        axes[i].set_visible(False)

    plt.tight_layout()
    plt.show()

    # Return results of normality tests as a DataFrame
    return pd.DataFrame(results).T

# Analyze distributions for weekly sales data
normality_results = plot_distribution_and_test_normality(weekly_sales_data, weekly_sales_data.columns)
print(normality_results)
```

## 4. Hypothesis Testing Between Two Weeks of Sales

### Purpose
This analysis aims to determine whether there is a statistically significant difference in the average sales between the first week (`W0`) and the last week (`W51`) of the dataset.

### Hypothesis
- **Null Hypothesis (H0)**: There is no difference in the mean sales between week 0 and week 51.
- **Alternative Hypothesis (H1)**: There is a difference in the mean sales between week 0 and week 51.

### Method
An independent t-test is used for this hypothesis test. This test compares the means of two independent groups (sales data from two different weeks in this case) to determine if there is statistical evidence that the associated population means are significantly different.

### Results Interpretation
- **T-statistic**: Measures the size of the difference relative to the variation in the sample data. Higher absolute values of the t-statistic indicate a greater difference between the groups.
- **P-value**: Provides the probability of seeing the effects observed in your sample under the null hypothesis. A p-value less than 0.05 typically indicates strong evidence against the null hypothesis, leading us to reject it.

```python
from scipy.stats import ttest_ind

# Assuming df is your DataFrame and has been previously loaded with weekly sales data
weekly_sales_data = df.loc[:, 'W0':'W51']

# Extract data for two weeks to compare
data_week1 = weekly_sales_data['W0']
data_week2 = weekly_sales_data['W51']

# Perform independent t-test
t_statistic, p_value = ttest_ind(data_week1, data_week2)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)
```

### Conclusion
Based on the p-value, if it is less than 0.05, we reject the null hypothesis, suggesting that the average sales in the two weeks are significantly different. If the p-value is greater than 0.05, we fail to reject the null hypothesis, indicating that the differences in weekly sales could be due to random chance rather than an actual difference in the means.

Data: https://archive.ics.uci.edu/dataset/368/facebook+metrics

This dataset contains data extracted from a Facebook page, focusing on various metrics associated with each post. The dataset includes details such as the type of content, whether the post was paid or not, engagement metrics like likes, shares, and comments, as well as reach and impression statistics.

Instruction: Choose a dataset and perform a data storytelling, choose the proper data visualization, do not use words, your graphs must tell a story. 

```python
import pandas as pd

# Load the dataset from the specified file
df = pd.read_csv('dataset_Facebook.csv', delimiter=';')  # Ensure the correct delimiter is used

print("First few rows of the dataset:")
display(df.head()) 
```

## Engagement Over Time by Month

This line chart shows the total engagement (likes, comments, shares) each month. It helps us identify when users are most active on the page, potentially guiding when to post content to maximize visibility and interaction.

```python
import matplotlib.pyplot as plt

monthly_engagement = df.groupby('Post Month')[['like', 'share', 'comment']].sum()
plt.figure(figsize=(10, 6))
monthly_engagement.plot(kind='line', marker='o')
plt.title('Engagement Over Time by Month')
plt.xlabel('Month')
plt.ylabel('Total Engagement')
plt.grid(True)
plt.show()
```

## Impact of Paid Promotion on Reach and Engagement

The bar chart compares the average reach and engagement between posts that were promoted through paid advertising and those that were not. This visualization highlights the effectiveness of paid promotions in increasing the visibility and interaction levels of the posts.

```python
import matplotlib.pyplot as plt

paid_engagement = df.groupby('Paid')[['Lifetime Post Total Reach', 'Lifetime Engaged Users']].mean()
fig, ax = plt.subplots(figsize=(10, 6))
paid_engagement.plot(kind='bar', ax=ax)
ax.set_xlabel('Paid Promotion (0 = No, 1 = Yes)')
ax.set_ylabel('Average Value')
plt.title('Impact of Paid Promotion on Reach and Engagement')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

## Content Type Performance

This bar chart demonstrates the effectiveness of different types of content (e.g., Photo, Status) based on their average reach and engagement. It provides insights into which content types are most engaging, helping to tailor future content strategies.


```python
import matplotlib.pyplot as plt

type_performance = df.groupby('Type')[['Lifetime Post Total Reach', 'Lifetime Engaged Users']].mean()
fig, ax = plt.subplots(figsize=(10, 6))
type_performance.plot(kind='bar', ax=ax)
ax.set_xlabel('Content Type')
ax.set_ylabel('Average Value')
plt.title('Content Type Performance')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Total Interactions by Post Hour

The bar chart shows the total interactions by hour of the day, indicating when posts achieve the highest levels of user engagement. This graph is crucial for planning the timing of posts to maximize visibility and interaction, illustrating the best hours to engage with the audience.


```python
import matplotlib.pyplot as plt

# Assuming df is the DataFrame already loaded
hourly_interactions = df.groupby('Post Hour')['Total Interactions']. sum()
plt.figure(figsize=(10, 6))
hourly_interactions.plot(kind='bar')
plt.title('Total Interactions by Post Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Interactions')
plt.grid(True)
plt.show()
```
